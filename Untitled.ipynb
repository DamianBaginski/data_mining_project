```python
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import LSTM, Dense, Embedding, ReLU, Dropout
import matplotlib.pyplot as plt
from keras.losses import BinaryCrossentropy
import seaborn as sns
from imblearn.over_sampling import SMOTE
from collections import Counter
from sklearn.metrics import roc_curve, auc

pd.set_option('display.max_columns', 999)
pd.set_option('display.max_rows', 10)
```

```python
data = pd.read_csv('aug_train.csv', sep=';')
print(data['experience'].unique())
data.head()
```

```python
data.isna().sum()
```

```python
print(data.shape)
data = data.dropna(subset=['gender', 'education_level', 'last_new_job', 'enrolled_university', 'experience'])
print(data.shape)
data = data.drop(columns=['company_size', 'company_type'])
print(data.shape)
```

```python
data.isna().sum()
```

```python
data['major_discipline'] = data['major_discipline'].fillna('No Major')
data['major_discipline'].unique()
```

```python
data.isna().sum()
```

```python
def replace_val_in_df(df: pd.DataFrame, col: str, what, to_what) -> pd.DataFrame:
    df[col] = df[col].replace(what, to_what)
    return df
```

```python
data = replace_val_in_df(data, 'relevent_experience', 'Has relevent experience', 1)
data = replace_val_in_df(data, 'relevent_experience', 'No relevent experience', 0)
data = replace_val_in_df(data, 'education_level', 'Primary School', 'Before College')
data = replace_val_in_df(data, 'education_level', 'High School', 'Before College')
data = replace_val_in_df(data, 'major_discipline', 'Arts', 'H&A')
data = replace_val_in_df(data, 'major_discipline', 'Humanities', 'H&A')

data['relevent_experience'].head()
```

```python
data['is_very_experienced'] = data.apply(lambda x: 1 if x['experience'] == '>20' else 0, axis=1)
data['is_not_experienced'] = data.apply(lambda x: 1 if x['experience'] == '<1' else 0, axis=1)

data.head()
```

```python
data_dum = data.drop(columns=['experience'])
```

```python
data_dum = pd.get_dummies(data_dum)
data_dum.head()
```

```python
y=data_dum['target']
X=data_dum.drop(columns=['target'])
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1)
X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, test_size=0.25, random_state=1)
```

```python
# X_train, y_train = SMOTE(random_state=42, sampling_strategy=0.3).fit_resample(X_train, y_train)
# print(sorted(Counter(y_train).items()))
```

```python
model = Sequential()
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
model.compile(optimizer='adam', loss=BinaryCrossentropy(), metrics=[tf.keras.metrics.Precision(), tf.keras.metrics.Recall(), tf.keras.metrics.AUC()])
```

```python
history = model.fit(X_train, y_train, epochs=100, batch_size=64, validation_data=(X_test, y_test), shuffle=False)
```

```python
plt.plot(history.history['precision_36'], label='train')
plt.plot(history.history['val_precision_36'], label='test')
plt.xlabel('Epochs')
plt.ylabel('Precision')
plt.legend()
```

```python
plt.plot(history.history['recall_22'], label='train')
plt.plot(history.history['val_recall_22'], label='test')
plt.xlabel('Epochs')
plt.ylabel('Recall')
plt.legend()
```

```python
plt.plot(history.history['auc_21'], label='train')
plt.plot(history.history['val_auc_21'], label='test')
plt.xlabel('Epochs')
plt.ylabel('ROC Index')
plt.legend()
```

```python
true_labels = y_val
binary_predictions = model.predict(X_val).flatten()
fpr, tpr, thresholds = roc_curve(true_labels, binary_predictions)

# Calculate the area under the ROC curve (AUC)
roc_auc = auc(fpr, tpr)

# Plot the ROC curve
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (AUC = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc="lower right")
plt.show()
```
